```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def simular_PAS(p,n,reps):
  # Input:
  # p: Es la probabilidad de tomar el valor 1 en cada paso, es decir de subir 1. Es un valor entre 0 y 1, la probabilidad de tomar -1, o bajar 1, será 1-p.
  # n: Es el numero de movimientos o pasos que damos en cada trayectoria.
  # reps: Es el numero de trayectorias.

  trayectoria = np.random.choice(a=[1,-1],p=[p,1-p],size=(reps,n))

  # Uso esta función para crear cada trayectoria, a es los valores que puede tomar en cada movimiento.
  # p es la probabilidad de tomar el valor 1, y el 1-p la probabilidad de tomar el -1.
  # size me genera un array del tamaño que le indique, en este caso con resp = numero de filas y n = numero de columnas

  output = np.cumsum(trayectoria,axis=1)  # Suma cada trayectoria por filas de izquierda a derecha, lo que me da la posicion tras cada paso.

  return output  # Shape: (reps, n)
```

```python
PAS_1 = simular_PAS(0.5,500,5)  # Funcion para generar la primera imagen página 16

x = np.linspace(0,500,500)
plt.plot(x,PAS_1.T,alpha=0.6) # Imprimo las trayectorias, igual que al final, el alpha sirve para que se vean mejor donde se cortan las trayectorias, en lugar de superponerse unas a otras

# Los siguientes plot son para que quede bonita la salida

plt.title(f'Simulación de {5} Paseos Aleatorios ({500} pasos)') # Título
plt.axhline(y=0, color='black', linestyle='--') # Línea discontinua en el cero para ver cuanto se alejan, si tienden a quedarse cerca de cero, suben o bajan
plt.xlabel('Movimientos/Pasos')
plt.ylabel('Posición') # Nombre a los ejes
plt.grid(True, alpha=0.3) # Hace la cuadrícula de detrás y el alpha hace que sea más transparente para que no se confunda con las líneas

plt.show()
```

```python
# El siguiente paso es hacer los PAS con media y desviacion tipica teóricas
# Imagen 4, diapositiva 30
p=0.5

PAS_4 = simular_PAS(p,500,20)

x = np.linspace(0,500,500)
plt.plot(x,PAS_4.T,alpha=0.6)

mean_t = x*(2*p-1)  # Calculo media y desviacion típica teóricas
std_t = np.sqrt(x*4*p*(1-p))

plt.plot(x,mean_t,color='black',linestyle='--',label='Media Teórica') # Muestro la recta de la media teórica

# Uso la función plot.fill_between para el sombreado de las imagenes, de la cual he buscado la informacion en internet como pone en las indicaciones, los colores elegidos son los mismos que en las diapositivas
plt.fill_between(x,mean_t+std_t,mean_t-std_t,color='gray',alpha=0.5,label=r'$\pm 1\sigma$') # El comando usado en el label lo he buscado en internet para que quede bonito el +- en la leyenda
plt.fill_between(x,mean_t+2*std_t,mean_t-2*std_t,color='gray',alpha=0.25,label=r'$\pm 2\sigma$')

plt.legend()

plt.title(f'Simulación de {20} Paseos Aleatorios ({500} pasos)')
plt.xlabel('Movimientos/Pasos')
plt.ylabel('Posición')
plt.grid(True, alpha=0.3)

plt.show()
```

Como pequeño apunte, los 3 casos anteriores tienen los mismos comentarios ya que he usado el mismo código para todos ellos. Tambien aclarar que en este caso concreto se pide replicar las imagenes de las diapositivas, por ello he ido poniendo yo el valor de la probabilidad p, pero para un caso más general se podría hacer de manera interactiva para que el usuario vaya introduciendo las probabilidades.

```python
# Como ultimo paso vamos a visualizar la imagen de la diapositiva 35
p=0.5
x = np.linspace(0,500,500)
mean_t = x*(2*p-1)  #Primero calculamos la media teórica

plt.plot(x,mean_t,color='red',label='Media teórica')

n_rep= np.array([5,100,200,500,1000])

for i in n_rep:
  PAS_7 = simular_PAS(p,500,i)
  mean_pas = np.mean(PAS_7,axis=0)
  plt.plot(x,mean_pas,alpha=0.6,label=f'{i}')

plt.legend()

plt.title(f'Simulación de {5} Paseos Aleatorios ({500} pasos)')
```
